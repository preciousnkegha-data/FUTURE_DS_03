```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from textblob import TextBlob
from wordcloud import WordCloud
```

```python
# Load Data

df=pd.read_excel("Student_Satisfaction_Survey.xlsx")
```

```python
df
```

```python
# Standardize Column Names

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("/", "_")
```

```python
df
```

```python
# Clean Data

df = df.drop_duplicates()
df = df.dropna(how="all")
```

```python
# Fill Missing Text Field

df = df.fillna({
    "basic_course": "Unknown",
    "questions": "",
    "course_name": "Unknown"
})
```

```python
# Extract Numeric Satisfaction Score
#     from average/percentage column

df["average_percentage_clean"] = (
    df["average__percentage"]
    .astype(str)
    .str.extract(r"(\d+\.?\d*)")[0]
    .astype(float)
)
```

```python
# The Satisfaction Metric

df["satisfaction_score"] = df["average_percentage_clean"]
```

```python
# Mean Satisfaction Score

mean_satisfaction = df["satisfaction_score"].mean().round(2)
print("Mean Satisfaction Score:", mean_satisfaction)
```

```python
# Clean Text for NLP (Question Cloumn)

df["questions_clean"] = (
    df["questions"]
    .astype(str)
    .str.lower()
    .str.replace(r"[^a-z\s]", " ", regex=True)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)
```

```python
# Textblob Topics + Wordcloud

def extract_noun_phrases(text):
    if not text:
        return []
    return [phrase.lower() for phrase in TextBlob(text).noun_phrases]

all_phrases = []
for q in df["questions_clean"]:
    all_phrases.extend(extract_noun_phrases(q))
```

```python
# Fall Back: if no noun phrases

if len(all_phrases) == 0:
    all_phrases = " ".join(df["questions_clean"]).split()

word_text = " ".join(all_phrases)
```

```python
# Word Cloud

wordcloud = WordCloud(width=1400, height=600, background_color="white").generate(word_text)
plt.figure(figsize=(14,7))
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Word Cloud of Most Common Topics in Questions")
plt.show()
```

```python
# Total Feedback

total_feedback = df["total_feedback_given"].sum()
print("Total Feedback:", total_feedback)
```

```python
# Total Courses

total_courses = df["course_name"].nunique()
print("Total Courses:", total_courses)
```

```python
# Total Questions

total_questions = df["questions"].nunique()
print("Total Questions:", total_questions)
```

```python
# Total Configured

total_configured = df["total_configured"].sum()
print("Total Configured:", total_configured)
```

```python
# Top 3 Courses with the highest satisfaction

top3_courses = (
    df.groupby("course_name")["satisfaction_score"]
    .mean()
    .sort_values(ascending=False)
    .head(3)
)

print("Courses With Highest Satisfaction")
print(top3_courses)
```

```python
top3_courses_df = top3_courses.reset_index()

plt.figure(figsize=(8,5))
sns.barplot(
    x='satisfaction_score', 
    y='course_name', 
    hue='course_name',  
    data=top3_courses_df, 
    palette='viridis',
    legend=False  
)

plt.title("Top 3 Courses With Highest Satisfaction")
plt.xlabel("Average Satisfaction Score")
plt.ylabel("Course Name")
plt.tight_layout()  
plt.show()
```

```python
# Correlation: Course Category vs Satisfaction

df["basic_course_code"] = df["basic_course"].astype("category").cat.codes

corr = df["basic_course_code"].corr(df["satisfaction_score"])
print("\nCorrelation between Course Category and Satisfaction:", corr)
```

```python
# Rating Distribution

plt.figure(figsize=(8,5))
sns.histplot(df["satisfaction_score"], kde=True)
plt.title("Distribution of Satisfaction Scores")
plt.show()
```

```python
# Average Satisfaction by Course Category

plt.figure(figsize=(12,6))
sns.barplot(data=df, x="basic_course", y="satisfaction_score", estimator="mean")
plt.xticks(rotation=45,ha="right")
plt.title("Average Satisfaction by Course Category")
plt.show()
```

```python
#  Overall Average Weightage

weight_cols = ["weightage_1", "weightage_2", "weightage_3", "weightage_4", "weightage_5"]
df["average_weightage"] = df[weight_cols].mean(axis=1)
df["average_weightage"]
```

```python
# Department Average

dept_avg_weightage = (
    df.groupby("basic_course")["average_weightage"]
    .mean()
    .sort_values(ascending=False)
    .reset_index()
)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    data=dept_avg_weightage,
    x="average_weightage",
    y="basic_course",
    hue="basic_course",  
    palette="viridis",
    legend=False  
)
plt.title("Average Weightage by Department")
plt.xlabel("Average Weightage")
plt.ylabel("Department")
plt.show()
```

```python
# Heatmap of all Weightage Columns

plt.figure(figsize=(8,6))
sns.heatmap(df[weight_cols].corr(), annot=True)
plt.title("Correlation Between Weightage Metrics")
plt.show()
```
